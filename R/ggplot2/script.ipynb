# Introduction to ggplot2

```r
# basic plot

x <- seq(-pi,pi,0.1)
y <- sin(x)
plot(x,y,main="Graph of sin(x),cos(x)",xlab = "x",type = 'l',col = "red")
lines(x,cos(x),col = "blue")
```

```r
x <- seq(-pi,pi,0.1)
y <- sin(x)
plot(x,y,main="Graph of sin(x),cos(x)",xlab = "x",type = 'l',col = "red")
lines(x,cos(x),col = "blue")
# Adding a legend
legend("topleft",c("sin(x)","cos(x)"),fill = c("blue","red"))
```

```r
# Using ggplot2 package

library(ggplot2)
```

```r
mov  <- read.csv("moviesData.csv")
```

```r
mov
```

```r
ggplot(data = mov , mapping = aes(x = critics_score,y = audience_score)) + geom_point()
```

```r
ggsave("scatterplot.png")
```

```r
data(mtcars)
mtcars
```

```r
ggplot(data = mtcars , mapping = aes(x=mpg,y=wt)) + geom_point()
```

```r
ggsave("mtcar MPG VS WT scatter.jpeg")
```

```r
ggplot(data = mtcars , mapping = aes(x=mpg,y=wt,color = am)) + geom_point()
```

```r
ggplot(data = mov , mapping = aes(x = critics_score,y = audience_score,color = genre)) + geom_point()
```

```r
ggsave("critics_score,audience_score genre.jpeg")
```

```r
str(mov$mpaa_rating)
levels(mov$mpaa_rating)
```

```r
ggplot(data = mov,mapping = aes(mpaa_rating)) + geom_bar()
```

```r
ggplot(data = mov,mapping = aes(mpaa_rating)) + geom_boxplot()
```

```r
ggplot(data = mov,mapping = aes(mpaa_rating)) + geom_bar() + labs(y="Rating count",title = "Count of mpaa rating")
```

```r
ggplot(data = mov,mapping = aes(mpaa_rating , fill = genre)) + geom_bar() + labs(y="Rating count",title = "Count of mpaa rating")
```

```r
ggplot(data = mov,mapping = aes(x=runtime))+geom_histogram() + labs(y="Runtime count",title = "Count of Runtime")
```
