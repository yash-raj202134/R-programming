```r
# filter genre according to drama
movie_drama  <-  filter(mov,genre == "Drama")
movie_drama
```

```r
# More than one filter using logical operator.
mov_com_drama <- filter(mov,genre == "Comedy"| genre == "Drama")
mov_com_drama
```

```r
# Can also use like this  %in% and vector i.e. match function
mov_com_drama <- filter(mov,genre %in% c("Comedy","Drama"))
mov_com_drama
```

```r

myfilter  <- filter(mov,genre =="Comedy" & imdb_rating >= 7.5)
myfilter
```

```r
# arrange() 
# arrange movies list on the basis of incresing order of imdb_rating

movImdb  <- arrange(mov,imdb_rating)
movImdb
```

```r
# arrange movies list on the basis of increasing order of runtime

movRuntime  <- arrange(mov,runtime)
movRuntime
```

```r
# arrange movies list on the basis of decresing  order of imdb_rating

movImdb  <- arrange(mov,desc(imdb_rating))
movImdb
```

```r
# arrange on the basis of genre and imdb_rating

mov_genre_imd  <- arrange(mov,genre,imdb_rating)
mov_genre_imd
```

```r
# Basics of pipe operator.



# Summarise () - reduces a data frame to a single row
# it summarises like mean ,median, etc. of the variable available in the data frame.
# we use summarise alog with group_by()
```

```r
summarise(mov,mean(imdb_rating))
```

```r
summarise(mov,median(imdb_rating))
```

```r
summarise(mov,mode(imdb_rating))
```

```r
# group_by()

groupMov <- group_by(mov,genre)
summarise(groupMov,mean(imdb_rating))
```

```r
dramaMov <- filter(mov,genre == "Drama")
gr_drama  <- group_by(dramaMov,mpaa_rating)
```

```r
summarise(gr_drama,mean(imdb_rating))
```

```r
# One liner :
summarise(group_by(filter(mov,genre == "Drama"),mpaa_rating),mean(imdb_rating))
```

```r
# Alternate method using pipe operator  %>%
# it prevents us making unnecessary data frames
# we can read the pipe as a series of imperative statement


mov %>% filter(genre == "Drama") %>% group_by(mpaa_rating) %>% summarise(mean(imdb_rating))
```

```r
library("ggplot2")
```

```r
# plots


mov %>% mutate(diff = audience_score - critics_score) %>% ggplot(mapping = aes(x = genre,y = diff)) + geom_boxplot()
```
