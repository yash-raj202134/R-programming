```r
# Declare a variable to read and store moviesData  
movies <- read.csv("moviesData.csv")
```

```r
# View the dimension of the data frame 
dim(movies)
```

```r
View(movies)
```

```r
# Plot a histogram using hist() function 

hist(movies$runtime)
```

```r
# tweeks in plot

hist(movies$runtime,
     main="Distribution of movies length",
     col="green",
     xlab="Runtime of movies -" ,
     xlim = c(0,300),
     ylim=c(0,300),
     breaks=4)
```

```r
# Pie charts plot

# create a table for movies genre
genreCount <- table(movies$genre)
genreCount

pie(genreCount,main = "Proportion of movies genre",border = "red")
```

```r
# Read the file moviesData.csv. Create a histogram of the object named imdb_num_votes in this file.
hist(movies$imdb_num_votes,main = "Distribution of movie imdb number of votes",xlab = "-imdb_num_votes-",xlim = c(0,10e+05),)
```

```r
# Create a pie chart of the object mpaarating.

mpaTable <- table(movies$mpaa_rating)
mpaTable
pie(mpaTable,main = "Proportion of MPAA rating")
```
