```r
# Declare a variable to read and store moviesData  
movies <- read.csv("moviesData.csv")
```

```r
moviesSub <- movies[1:20,]
moviesSub
```

```r
# Barchart plot

barplot(moviesSub$imdb_rating,main="Movies IMDB rating",ylab = "IMDB rating",ylim = c(0,10),
        names.arg = moviesSub$title,
       las=2)
```

```r
# Scatter plot
# imdb rating vs audience score

plot(x=movies$imdb_rating,
     y=movies$audience_score,
     main="IMDB rating vs Audience score",
     xlab = "IMDB rating",
     ylab = "Audience Score",
     xlim = c(0,9),
     ylim = c(0,99),
     col = "red"
     )
```

```r
# Correlation cor()
cor(movies$imdb_rating,movies$audience_score)
```

```r
# Create a bar chart of critics score for the first 10movies.

moviesSub <- movies[1:10,]

barplot(moviesSub$critics_score,main="Critics score",ylab = "Critics score",ylim = c(0,100),
        names.arg = moviesSub$title,
       las=2)
```

```r
# Create a scatter plot of imdb_rating and imdb_num_votes to see their relation.

plot(x = movies$imdb_rating,
    y = movies$imdb_num_votes,
    main="IMDB rating vs IMDB number of Votes",
    xlab="Imdb rating",
    ylab="Imdb no. of votes",
    xlim=c(0,10),
    col="red",)

cor(movies$imdb_rating,movies$imdb_num_votes)
```
